```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

# 1 Web scraping

From website https://index.minfin.com.ua/ua/reference/coronavirus/2020-02/

I took the data between the dates 05.2020 and 12.2021

```python
years = ['2020', '2021']
months = ['01','02','03','04','05','06','07','08','09','10','11','12']
```

```python
covid_19_in_Ukraine_full_time = []
```

```python
# collect data

for year in years:
    for month in months:
        if year == '2020' and month in ['01','02','03','04']:
            continue

        url = f"https://index.minfin.com.ua/ua/reference/coronavirus/ukraine/{year}-{month}/"

        r = requests.get(url)       
        soup = BeautifulSoup(r.text, 'lxml')

        loc = 0
        step = 4

        while True:
            month_data = []

            for el in soup.findAll('div', class_ = "compact-table")[1].findAll('td')[loc:loc+step]:
                month_data.append(el.text)

            if len(month_data) == 0 or 'Усього' in month_data:
                break
            loc += step

            covid_19_in_Ukraine_full_time.append(month_data)
```

```python
df = pd.DataFrame(columns = ['date', 'total_infected', 'deadly_cases', 'recovered'], data = covid_19_in_Ukraine_full_time)
```

```python
df
```

```python
covid_19_Ukraine_df = pd.DataFrame(columns = ['date','day_week', 'total_infected', 'deadly_cases', 'recovered'])
```

```python
# split column 'date' into two columns 'date' and 'day_week'
day_week = []
date = []

for element in df.date:
    day_week.append(element[11:])
    date.append(element[:11])
    
    
covid_19_Ukraine_df['date'] = date
covid_19_Ukraine_df['day_week'] = day_week
```

```python
# convert columns [total_infected, deadly_cases, recovered] into int type
for column in df.columns.drop('date', 1):
    covid_19_Ukraine_df[column] = df[column].values.astype('int')
```

```python
covid_19_Ukraine_df['date'] = pd.to_datetime(covid_19_Ukraine_df['date'])
```

# 2.2 Statistics by days

```python
covid_19_Ukraine_df_statistics_day = pd.DataFrame(columns=['columns','max','date(max)','min','date(min)'])
```

```python
# getting  statistics by deys - min,max 
for column in covid_19_Ukraine_df.drop(['date','day_week'], 1).columns:
    covid_19_Ukraine_df_statistics_day = covid_19_Ukraine_df_statistics_day.append({
                                        'columns' : column,
                                        'max' : covid_19_Ukraine_df[column].max(),
                                        'date(max)' : covid_19_Ukraine_df.loc[covid_19_Ukraine_df[column].max() == covid_19_Ukraine_df[column]].date.astype('str').values[0],
                                        'min' : covid_19_Ukraine_df[column].min(),
                                        'date(min)' : covid_19_Ukraine_df.loc[covid_19_Ukraine_df[column].min() == covid_19_Ukraine_df[column]].date.astype('str').values[0]}, 
                                        ignore_index=True)
```
